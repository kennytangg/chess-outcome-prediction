### 3. Drop columns that has many missing data or irrelevant

```python
df = df.drop(['Site', 'Round', 'PlyCount', 'EventDate'], axis=1)
df = df.dropna(subset=['ECO', 'Moves'])
df.isnull().sum()
```

### 4. Convert data types

```python
df['WhiteElo'] = df['WhiteElo'].astype(int)
df['BlackElo'] = df['BlackElo'].astype(int)
df.info()
```

### 5. Remove * or unfinished games rows 

```python
df = df[df['Result'].isin(['1-0', '0-1', '1/2-1/2'])]
df['Result'].value_counts()
```

### 6. Extract termination feature

```python
def extract_termination(term_str):
    term_str = str(term_str).lower()
    
    if any(word in term_str for word in ['rules', 'infraction', 'cheat', 'violation']):
        return 'violation'
    
    elif any(word in term_str for word in ['time', 'forfeit', 'outoftime']):
        return 'time'
    
    elif any(word in term_str for word in ['abandon', 'unterminated', 'disconnect']):
        return 'abandon'
    
    else:
        return 'normal'

df['TerminationType'] = df['Termination'].apply(extract_termination)
df = df[df['TerminationType'].isin(['normal', 'time'])]
df['TerminationType'].value_counts()
```

### 7. Parse and group game type into 3 categories

```python
#df['Event'].unique()
df['Event'].value_counts()
```

```python
def extract_time_class(event):
    event_lower = str(event).lower()
    if 'blitz' in event_lower:
        return 'blitz'
    elif 'rapid' in event_lower:
        return 'rapid'
    elif 'classical' in event_lower:
        return 'classical'
    else:
        return np.nan

df['EventType'] = df['Event'].apply(extract_time_class)
df = df[df['EventType'].notna()]
df['EventType'].value_counts(dropna=False)
```

### 8. Drop duplicates 

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

### 9. Extract move features

```python
def extract_full_game_features(moves_str):
    """Extract interpretable features from complete move sequence"""
    if pd.isna(moves_str) or moves_str == "":
        return {f: 0 for f in [
            'total_moves', 'w_captures', 'w_checks', 'w_castles', 'w_promotions',
            'b_captures', 'b_checks', 'b_castles', 'b_promotions',
            'w_capture_rate', 'b_capture_rate', 'w_aggression_score', 'b_aggression_score'
        ]}
    
    moves = moves_str.split()
    total_moves = len(moves)
    half_moves_w = (total_moves + 1) // 2
    half_moves_b = total_moves // 2
    
    w_captures = sum(1 for i, m in enumerate(moves) if i % 2 == 0 and 'x' in m)
    w_checks = sum(1 for i, m in enumerate(moves) if i % 2 == 0 and '+' in m)
    w_castles = sum(1 for i, m in enumerate(moves) if i % 2 == 0 and 'O-O' in m)
    w_promotions = sum(1 for i, m in enumerate(moves) if i % 2 == 0 and '=' in m)
    
    b_captures = sum(1 for i, m in enumerate(moves) if i % 2 == 1 and 'x' in m)
    b_checks = sum(1 for i, m in enumerate(moves) if i % 2 == 1 and '+' in m)
    b_castles = sum(1 for i, m in enumerate(moves) if i % 2 == 1 and 'O-O' in m)
    b_promotions = sum(1 for i, m in enumerate(moves) if i % 2 == 1 and '=' in m)
    
    return {
        'total_moves': total_moves,
        'w_captures': w_captures,
        'w_checks': w_checks,
        'w_castles': w_castles,
        'w_promotions': w_promotions,
        'b_captures': b_captures,
        'b_checks': b_checks,
        'b_castles': b_castles,
        'b_promotions': b_promotions,
        'w_capture_rate': w_captures / max(1, half_moves_w),
        'b_capture_rate': b_captures / max(1, half_moves_b),
        'w_aggression_score': (w_captures + w_checks + w_promotions) / max(1, half_moves_w),
        'b_aggression_score': (b_captures + b_checks + b_promotions) / max(1, half_moves_b),
    }

print("Extracting move features...")
move_features = df['Moves'].apply(extract_full_game_features)
move_features_df = pd.DataFrame(move_features.tolist())
```

### 10. Combine all features

```python
df = pd.concat([df.reset_index(drop=True), move_features_df.reset_index(drop=True)], axis=1)
df.head()
```

### 12. Export cleaned data

```python
df.to_csv('data/chess_games_clean.csv', index=False)
```
